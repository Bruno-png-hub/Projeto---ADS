```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados simulados com base em estudos citados
generos = [
    "Música Clássica",
    "Lo-fi Hip Hop",
    "Jazz Suave",
    "Natureza / Sons Ambientes",
    "Música Eletrônica Suave",
    "Pop Instrumental",
    "Rock Pesado",
    "Silêncio"
]

# Percentual de eficácia na concentração com base em estudos/recomendações
eficacia = [85, 80, 75, 70, 65, 60, 35, 50]

# Cores: tons frios para música eficaz e quente para menos eficaz
cores = ['#4caf50', '#66bb6a', '#81c784', '#aed581', '#dce775', '#fff176', '#ef5350', '#90a4ae']

# Criando o gráfico
plt.figure(figsize=(10, 6))
bars = plt.barh(generos, eficacia, color=cores)
plt.xlabel("Eficácia na Concentração (%)")
plt.title("Gêneros Musicais e sua Eficácia na Concentração")
plt.xlim(0, 100)

# Exibir valores nas barras
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width()}%', va='center')

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
