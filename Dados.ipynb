```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados simulados com base em estudos citados
generos = [
    "Msica Clssica",
    "Lo-fi Hip Hop",
    "Jazz Suave",
    "Natureza / Sons Ambientes",
    "Msica Eletrnica Suave",
    "Pop Instrumental",
    "Rock Pesado",
    "Silncio"
]

# Percentual de eficcia na concentrao com base em estudos/recomendaes
eficacia = [85, 80, 75, 70, 65, 60, 35, 50]

# Cores: tons frios para msica eficaz e quente para menos eficaz
cores = ['#4caf50', '#66bb6a', '#81c784', '#aed581', '#dce775', '#fff176', '#ef5350', '#90a4ae']

# Criando o grfico
plt.figure(figsize=(10, 6))
bars = plt.barh(generos, eficacia, color=cores)
plt.xlabel("Eficcia na Concentrao (%)")
plt.title("Gneros Musicais e sua Eficcia na Concentrao")
plt.xlim(0, 100)

# Exibir valores nas barras
for bar in bars:
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width()}%', va='center')

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
