```python

# Projeto: Efeitos da Música na Concentração
# Tema: Classificação e análise de gêneros musicais usando GTZAN Dataset
# Plataforma: Google Colab

!pip install librosa pandas matplotlib seaborn --quiet

import librosa
import librosa.display
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()
import zipfile
import shutil


zip_path = "/content/archive.zip"
extract_to = "/content/genres"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

import os

# Verificar subpastas dentro de /content/genres
print(os.listdir("/content/genres"))


# Configuração para melhor visualização
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Caminho da pasta com os arquivos de áudio
DATA_PATH = '/content/genres'

# ✅ Garante que a pasta exista
os.makedirs(DATA_PATH, exist_ok=True)

# Função para extrair características de um único arquivo de áudio
def extract_features(file_path):
    try:
        y, sr = librosa.load(file_path, duration=30)
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
        chroma = librosa.feature.chroma_stft(y=y, sr=sr)
        zcr = librosa.feature.zero_crossing_rate(y)
        spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr)

        features = {
            'mfcc_mean': np.mean(mfcc),
            'chroma_mean': np.mean(chroma),
            'zcr_mean': np.mean(zcr),
            'spectral_centroid_mean': np.mean(spectral_centroid)
        }
        return features
    except Exception as e:
        print("Erro ao processar", file_path)
        return None

# Coletar características de todos os arquivos
genres = os.listdir(DATA_PATH)
data = []

for genre in genres:
    genre_path = os.path.join(DATA_PATH, genre)
    if not os.path.isdir(genre_path): continue
    for file in os.listdir(genre_path):
        if file.endswith('.wav'):
            file_path = os.path.join(genre_path, file)
            features = extract_features(file_path)
            if features:
                features['genre'] = genre
                data.append(features)

# Criar DataFrame
df = pd.DataFrame(data)

# Mostrar primeiras linhas
df.head()

df.to_csv("gtzan_features.csv", index=False)



```
